```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import scipy.io as scio
import h5py
%matplotlib inline
```

```python
def myImgShow(img):      #将opencv对象转为plt对象，以显示在jupyter上
    show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(show_img)
    plt.show()
```

```python
side = 11 #分割成多少份
eachCellNum = 6 #每个网格标注数据占用数目

tmpList = np.zeros((30000,726))

imgFile = "/media/lgh/Passport/sharefolder/CelebA/Img/img_celeba.7z/img_celeba/"

labelFile = "/media/lgh/Passport/sharefolder/CelebA/Anno/list_bbox_celeba.txt"
file = open(labelFile,mode='r')
file.readline()
file.readline() #读取前两行无用数据

count = 0

print(tmpList.shape)
for i in range(30000):
    count = count +1
    img = cv2.imread(imgFile+"%06d.jpg"%(i+1))
    imgH = img.shape[0]
    imgW = img.shape[1]
    channel = img.shape[2] #读取图片尺寸
    
    temp = file.readline().split()[1:]  #读取框的x,y,w,h
    for itemp in range(4):
        temp[itemp] = int(temp[itemp])
    x = (temp[0] + 0.5*temp[2])/imgW
    y = (temp[1] + 0.5*temp[3])/imgH
    
    col = int(x*side) 
    row = int(y*side)   #中心店的行数和列数，从0开始算
    
    cell_num = row*side + col #网格序号，从0开始算
    
    w = temp[2]/imgW
    h = temp[3]/imgH  #计算相对高度、宽度
    
    tmpList[i,cell_num*eachCellNum] = 1 #confidence
    tmpList[i,cell_num*eachCellNum+1] = 1 #class
    #tmpList[i,cell_num*eachCellNum+2] = x*side - col
    #tmpList[i,cell_num*eachCellNum+3] = y*side - row
    tmpList[i,cell_num*eachCellNum+2] = x*side
    tmpList[i,cell_num*eachCellNum+3] = y*side
    
    tmpList[i,cell_num*eachCellNum+4] = w*side
    tmpList[i,cell_num*eachCellNum+5] = h*side
    print ("count ",count)
    if count > 29998:
#在图片显示中心所在网格和中心点，直观验证数据正确性    
        cv2.circle(img,(int(x*imgW),int(y*imgH)),10,(0,0,250),-1)
        cv2.circle(img,(int(col*imgW/side),int(row*imgH/side)),10,(0,0,250),-1)
        cv2.circle(img,(int((col+1)*imgW/side),int(row*imgH/side)),10,(0,0,250),-1)
        cv2.circle(img,(int((col+1)*imgW/side),int((row+1)*imgH/side)),10,(0,0,250),-1)
        cv2.circle(img,(int(col*imgW/side),int((row+1)*imgH/side)),10,(0,0,250),-1)
        
        red = (0,0,255)
        cv2.line(img, (temp[0],temp[1]),(temp[0]+temp[2],temp[1]),red,10)
        cv2.line(img, (temp[0],temp[1]+temp[3]),(temp[0]+temp[2],temp[1]+temp[3]),red,10)
        cv2.line(img, (temp[0],temp[1]),(temp[0],temp[1]+temp[3]),red,10)
        cv2.line(img, (temp[0]+temp[2],temp[1]),(temp[0]+temp[2],temp[1]+temp[3]),red,10)
        
        myImgShow(img)
```
